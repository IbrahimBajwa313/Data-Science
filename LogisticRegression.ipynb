# Task 2: Feature Selection Techniques

### 1. Correlation-Based Feature Selection

##### First, we will calculate the correlation matrix and remove one of each pair of highly correlated features (threshold > 0.85).

```python
# Compute correlation matrix
corr_matrix = df.corr().abs()

# Set a threshold for high correlation
threshold = 0.85

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python

# Identify columns with high correlation
columns_to_drop = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > threshold:
            columns_to_drop.add(corr_matrix.columns[i])

# Drop the correlated columns
df.drop(columns=columns_to_drop, inplace=True)

# Display dropped columns
print(f"Columns dropped due to high correlation: {columns_to_drop}")
```

### 2. Recursive Feature Elimination (RFE)

##### Next, we will use RFE with Logistic Regression to select the top k most important features.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Define the feature set and target variable
X = df.drop('satisfaction', axis=1)
y = df['satisfaction']

# Create a Logistic Regression model
model = LogisticRegression(max_iter=5000)

# Perform RFE
selector = RFE(model, n_features_to_select=15)  
selector = selector.fit(X, y)

# Get selected features
selected_features_rfe = X.columns[selector.support_]
print("Selected features using RFE:", selected_features_rfe.tolist())
```

```python
# Train Logistic Regression on selected features (RFE)
X_selected_rfe = X[selected_features_rfe]  # Using selected features from RFE
X_train_rfe, X_test_rfe, y_train, y_test = train_test_split(X_selected_rfe, y, test_size=0.2, random_state=42)

model_rfe = LogisticRegression(max_iter=5000, solver='saga')
model_rfe.fit(X_train_rfe, y_train)

# Predictions using RFE features
y_pred_rfe = model_rfe.predict(X_test_rfe)

# Evaluate model using RFE features
metrics_rfe = {
    "Accuracy": accuracy_score(y_test, y_pred_rfe),
    "Precision": precision_score(y_test, y_pred_rfe, average='weighted'),
    "Recall": recall_score(y_test, y_pred_rfe, average='weighted'),
    "F1 Score": f1_score(y_test, y_pred_rfe, average='weighted')
}
```

### 3. SelectKBest (Statistical Approach)

##### Finally, we will use SelectKBest with the ANOVA F-test to select the best features.

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Apply SelectKBest
selector = SelectKBest(score_func=f_classif, k=10) 
X_new = selector.fit_transform(X, y)

# Get the selected feature indices
selected_indices = selector.get_support(indices=True)
selected_features_kbest = X.columns[selected_indices]

print("Selected features using SelectKBest:", selected_features_kbest.tolist())
```

```python
### **Summary of Feature Selection Results**
summary = {
    "Removed (Correlation-Based)": list(columns_to_drop),
    "Selected (RFE)": selected_features_rfe.tolist(),
    "Selected (SelectKBest)": selected_features_kbest.tolist()
}

print("\nFeature Selection Summary:")
for key, value in summary.items():
    print(f"{key}: {value}")
```

# Task 3: Train Logistic Regression Model

```python
# Train Logistic Regression on selected features (SelectKBest)
X_selected_kbest = X[selected_features_kbest]  # Using selected features from SelectKBest
X_train_kbest, X_test_kbest, y_train, y_test = train_test_split(X_selected_kbest, y, test_size=0.2, random_state=42)

model_kbest = LogisticRegression(max_iter=5000, solver='saga')
model_kbest.fit(X_train_kbest, y_train)

# Predictions using SelectKBest features
y_pred_kbest = model_kbest.predict(X_test_kbest)

# Evaluate model using SelectKBest features
metrics_kbest = {
    "Accuracy": accuracy_score(y_test, y_pred_kbest),
    "Precision": precision_score(y_test, y_pred_kbest, average='weighted'),
    "Recall": recall_score(y_test, y_pred_kbest, average='weighted'),
    "F1 Score": f1_score(y_test, y_pred_kbest, average='weighted')
}
```

```python
# Print performance comparison
print("\nModel Performance Comparison:")
print("\n Using All Features:", metrics_all)
print("\n Using SelectKBest Features:", metrics_kbest)
print("\n Using RFE Features:", metrics_rfe)
```

# Task 4: Model Performance Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc

# Function to evaluate a model
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probabilities for ROC curve

    # Compute performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # ROC Curve and AUC Score
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    auc_score = auc(fpr, tpr)

  # Print Metrics
    print(f"\n Model Performance: {model_name}")
    print(f" Accuracy: {accuracy:.4f}")
    print(f" Precision: {precision:.4f}")
    print(f" Recall: {recall:.4f}")
    print(f" F1 Score: {f1:.4f}")
    print(f" AUC Score: {auc_score:.4f}")

  # Plot Confusion Matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

    # Plot ROC Curve
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, color='blue', label=f"AUC = {auc_score:.4f}")
    plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name}")
    plt.legend()
    plt.show()

  
```

```python

# Evaluate models
evaluate_model(model_all, X_test_all, y_test, "All Features")
evaluate_model(model_kbest, X_test_kbest, y_test, "SelectKBest Features")
evaluate_model(model_rfe, X_test_rfe, y_test, "RFE Features")
```
