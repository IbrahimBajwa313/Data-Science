#  Importing data and preprocessing

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\SMART TECH\Desktop\sem6\Breast Cancer Dataset.csv")
```

```python
# Handle missing values if any (e.g., drop or impute)
# Check for missing Values 
print(df.isnull().sum())
 
```

```python
df = df.drop(columns = ["Unnamed: 32"])

# Drop irrelevant columns
df.drop(columns=['id'], inplace=True)

 # inplace=True parameter means the operation alters the original DataFrame rather than creating a new one.
```

```python
df
```

```python
# Remove duplicate rows if any
df.drop_duplicates(inplace=True)
```

```python
# Encode categorical variables (e.g., 'diagnosis': M=1, B=0)

df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df
```

```python
# Separate features and targetfrom sklearn.feature_selection import mutual_info_classif
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

# Standarize the features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_scaled = pd.DataFrame(X_scaled , columns=X.columns) 
```

# Task 1. Compute feature importance using mutual information

```python
from sklearn.feature_selection import mutual_info_classif

# Compute MI Score
mi_score = mutual_info_classif(X_scaled,y, random_state=42)

mi_df = pd.DataFrame({'feature': X_scaled.columns, 'MI Score':mi_score})
mi_df = mi_df.sort_values(by='MI Score', ascending=False)

mi_df
```

# 3. Visualize feature importance

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.barh(mi_df['feature'] , mi_df['MI Score'])
plt.xlabel('MI Score')
plt.ylabel('Features')

plt.title('Feature Importance using Mutual Information')
plt.gca().invert_yaxis()
plt.show()
```

# 4. Identify top 5 features and train a model

```python
top_5_features = mi_df['feature'].head(5).values
X_top5 = X_scaled[top_5_features]
X_top5
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X_top5, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluate Model
y_predict = model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)

print(f"Top 5 Features: {top_5_features}")
print(f"Model Accuracy using Top 5 Features: {accuracy:.4f}")
```

# Task 2: Applying Variance Threshold for Feature Selection

```python
from sklearn.feature_selection import VarianceThreshold

# Applying Variance Threshold | Removing Near Zero Threshold Features
selector = VarianceThreshold( threshold = 0.1 )
x_reduced = selector.fit_transform(X)
```

```python
# Get the variance of each feature
feature_variances = selector.variances_

# Create a DataFrame to display feature names with their variance
variance_df = pd.DataFrame({'Feature': X.columns, 'Variance': feature_variances})

# Display the variance of each feature
print(variance_df)
```

```python
# Get Retained Featues names
# Get retained feature indices
retained_indices = selector.get_support(indices=True)
retained_features = X.columns[retained_indices] 
print("Retained Features:", retained_features)
```

```python
# Compare with original set
print(f"Number of features before: {X.shape[1]}, after: {len(retained_features)}")


# print(f"Original Feature Count: {X.shape[1]}, Retained Feature Count: {retained_features[1]}")
```
