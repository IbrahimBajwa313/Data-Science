#  Importing data and preprocessing

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\SMART TECH\Desktop\sem6\Breast Cancer Dataset.csv")
```

```python
# Handle missing values if any (e.g., drop or impute)
# Check for missing Values 
print(df.isnull().sum())
 
```

```python
df = df.drop(columns = ["Unnamed: 32"])

# Drop irrelevant columns
df.drop(columns=['id'], inplace=True)

 # inplace=True parameter means the operation alters the original DataFrame rather than creating a new one.
```

```python
df
```

```python
# Remove duplicate rows if any
df.drop_duplicates(inplace=True)
```

```python
# Encode categorical variables (e.g., 'diagnosis': M=1, B=0)

df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df
```

```python
# Separate features and targetfrom sklearn.feature_selection import mutual_info_classif
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

print(X.shape)  # Should print (569, n_features)
print(y.shape) 
```

# Standarize the features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_scaled = pd.DataFrame(X_scaled , columns=X.columns) 
```

# Task 1. Compute feature importance using mutual information

```python
from sklearn.feature_selection import mutual_info_classif

# Compute MI Score
mi_score = mutual_info_classif(X_scaled,y, random_state=42)

mi_df = pd.DataFrame({'feature': X_scaled.columns, 'MI Score':mi_score})
mi_df = mi_df.sort_values(by='MI Score', ascending=False)

mi_df
```

# 3. Visualize feature importance

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.barh(mi_df['feature'] , mi_df['MI Score'])
plt.xlabel('MI Score')
plt.ylabel('Features')

plt.title('Feature Importance using Mutual Information')
plt.gca().invert_yaxis()
plt.show()
```

# 4. Identify top 5 features and train a model

```python
top_5_features = mi_df['feature'].head(5).values
X_top5 = X_scaled[top_5_features]
X_top5
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X_top5, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluate Model
y_predict = model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)

print(f"Top 5 Features: {top_5_features}")
print(f"Model Accuracy using Top 5 Features: {accuracy:.4f}")
```

# Task 2: Applying Variance Threshold for Feature Selection

```python
from sklearn.feature_selection import VarianceThreshold

# Applying Variance Threshold | Removing Near Zero Threshold Features
selector = VarianceThreshold( threshold = 0.1 )
x_reduced = selector.fit_transform(X)
```

```python
# Get the variance of each feature
feature_variances = selector.variances_

# Create a DataFrame to display feature names with their variance
variance_df = pd.DataFrame({'Feature': X.columns, 'Variance': feature_variances})

# Display the variance of each feature
print(variance_df)
```

```python
# Get Retained Featues names
# Get retained feature indices
retained_indices = selector.get_support(indices=True)
retained_features = X.columns[retained_indices] 
print("Retained Features:", retained_features)
```

```python
# Compare with original set
print(f"Number of features before: {X.shape[1]}, after: {len(retained_features)}")


# print(f"Original Feature Count: {X.shape[1]}, Retained Feature Count: {retained_features[1]}")
```

# Task 3: Forward Feature Selection using Logistic Regression

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LogisticRegression
```

```python
# Initialize the model
model = LogisticRegression(max_iter=10000)

# Initialize SFS
sfs = SFS(model, 
          k_features='best', 
          forward=True, 
          floating=False, # When set to False, it performs a strict forward selection. If set to True, it allows for adding or removing features from the selected set, giving it more flexibility to find an optimal feature subset.
          scoring='accuracy', #  measures the proportion of correct predictions made by the model.
          cv=5) # cv=5 means that the dataset will be split into 5 parts. The model will be trained on 4 parts and tested on 1 part, and this process will be repeated 5 times.  

# Perform SFS
sfs = sfs.fit(X, y)

# Selected features
selected_features = list(sfs.k_feature_names_)
print("Selected Features:", selected_features)
```

```python
# Train model with selected features
X_sfs = X[selected_features]
X_train_sfs, X_test_sfs, y_train, y_test = train_test_split(X_sfs, y, test_size=0.3, random_state=0)

model_sfs = LogisticRegression(max_iter=10000)
model_sfs.fit(X_train_sfs, y_train)
y_pred_sfs = model_sfs.predict(X_test_sfs)

accuracy_sfs = accuracy_score(y_test, y_pred_sfs)
print(f"Accuracy with selected features: {accuracy_sfs:.2f}")
```

# Task 4: Recursive Feature Elimination (RFE) for Feature Selection

```python
from sklearn.feature_selection import RFE
```

```python
# Initialize RFE
rfe = RFE(estimator=model, n_features_to_select=7)
```

```python
# Fit RFE
rfe = rfe.fit(X, y)
```

```python
# Train model with selected features
X_rfe = X[selected_features_rfe]
X_train_rfe, X_test_rfe, y_train, y_test = train_test_split(X_rfe, y, test_size=0.3, random_state=0)
```

```python
# Fiting the model
model_rfe = LogisticRegression(max_iter=10000)
model_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = model_rfe.predict(X_test_rfe)
```

```python
accuracy_rfe = accuracy_score(y_test, y_pred_rfe)
print(f"Accuracy with RFE-selected features: {accuracy_rfe:.2f}")
```

# Task 5: Feature Importance using Random Forest

```python
# Split the dataset into training and testing sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Standardize the dataset (helps improve model performance)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=0)
rf_model.fit(X_train, y_train)

 
```

# Extract and Visualize Feature Importances

```python
# Extract feature importances
importances = rf_model.feature_importances_

# Create a DataFrame for visualization
rf_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
rf_importances = rf_importances.sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(12, 6))
plt.barh(rf_importances['Feature'], rf_importances['Importance'], color='skyblue')
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance using Random Forest")
plt.gca().invert_yaxis()  # Invert y-axis to display the most important feature on top
plt.show()
```

 # Remove the Least Important 3 Features and Retrain the Model

```python
# Remove the 3 least important features
least_important_features = rf_importances.tail(10)['Feature'].values
X_selected = X.drop(columns=least_important_features)

# Split dataset again
X_train_sel, X_test_sel, y_train_sel, y_test_sel = train_test_split(X_selected, y, test_size=0.2, random_state=0)

# Standardize the dataset
X_train_sel = scaler.fit_transform(X_train_sel)
X_test_sel = scaler.transform(X_test_sel)

# Retrain Random Forest with selected features
rf_model_sel = RandomForestClassifier(n_estimators=100, random_state=0)
rf_model_sel.fit(X_train_sel, y_train_sel)
```

```python
# Evaluate both models
y_pred_original = rf_model.predict(X_test)
y_pred_selected = rf_model_sel.predict(X_test_sel)

accuracy_original = accuracy_score(y_test, y_pred_original)
accuracy_selected = accuracy_score(y_test_sel, y_pred_selected)

print(f"Model Accuracy with All Features: {accuracy_original:.4f}")
print(f"Model Accuracy after Removing Least Important 3 Features: {accuracy_selected:.4f}")
```
