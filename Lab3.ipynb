```python
import pandas as pd

df = pd.read_csv(r"C:\Users\SMART TECH\Desktop\AppliedNLPMaterial-master\tested.csv")  
print(df.head(50))  # Display first 5 rows
```

#### Task 1: Impute missing values in the Age column using the median

```python
df['Age'] = df['Age'].fillna(df['Age'].median())  # Ensure proper dtype

print(df.head(50))
```

##### Task 2: Fill missing values in the Embarked column with the most frequent value



```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode())
print(df.head(50))
```

####### Task3: Drop rows with missing Cabin data

```python
df = df.dropna(subset = ['Cabin'])
print(df.head(50))
```

## Task4: Convert the Pclass column to a categorical variable

```python
df["Pclass"] = df["Pclass"].astype("category") 
print(df.head(50))
```

# Task5: Ensure the Fare column is a float, not an integer

```python
df.Fare = df.Fare.astype(float)
```

```python
print(df['Fare'].dtype)
df
```

# Task6: Analyze if Age has outliers and handle them accordingly

#### Detect outliers using the IQR method and remove them:

```python
Q1 = df.Age.quantile(0.25)
Q3 = df.Age.quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df.Age >= lower_bound) & (df.Age <= upper_bound)]
df
```

# Task7: Remove any duplicate rows if found

```python
df = df.drop_duplicates()
df
```

```python
print(df.dtypes)
from sklearn.preprocessing import LabelEncoder

# Convert categorical columns to numerical
label_encoder = LabelEncoder()

# Example: Convert 'Sex' and 'Embarked' to numbers
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
df['Pclass'] = df['Pclass'].astype(int)  # Convert to int if necessary
```

```python
df = df.drop(columns=['Name', 'Ticket', 'Cabin'])

print(df.dtypes)
```

```python
df = df.astype(float) 

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
 

# Selecting features and target variable (Assuming 'Survived' is the target variable)
x = df.drop(columns=['Survived'])  # Features (Independent Variables)
y = df['Survived']  # Target Variable

# Splitting data into 70% training and 30% testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# Initialize Logistic Regression model
lr = LogisticRegression()

# Train the model
lr.fit(x_train, y_train)

# Predict on the test set
y_pred = lr.predict(x_test)

# Compute Model Accuracy
accuracy = metrics.accuracy_score(y_test, y_pred) * 100
print("Accuracy:", accuracy)
```
